```python
def process_text(msg):
    """
    To process the text into a list of number 
    Parameters
    ----------
    msg : message
        This object represents a message
    Returns
    -------
    reshaped :2D numpy array 
      to use it for predictions  
    numbers_only : List
        list of numbers that are in the message 

    """
    message=msg.text #For text messages, the actual UTF-8 text of the message
    msg_lst=message.split('\n') 
    numbers_only = [float(item.split(":")[1]) for item in msg_lst] 
    reshaped=np.array((numbers_only)).reshape(1,-1)
    return reshaped,numbers_only
```
